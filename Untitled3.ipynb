```python
words = open('./names.txt','r').read().splitlines()
words = ['.'+w for w in words[:]]
words[-1] += '.'
words = ''.join(words)
chars1 = list(sorted(set(words)))
chars2 = list(sorted(set((i,k) for i,k in zip(words,words[1:-1]))))
len(chars2)
```

```python
chars2
```

```python
c1toi = {chars1[i]:i for i in range(len(chars1))}
c2toi = {chars2[i]:i for i in range(len(chars2))}

itoc1 = {i:x for x,i in c1toi.items()}
itoc2 = {i:(a,b) for (a,b),i in c2toi.items()}
```

```python
import torch
import torch.nn.functional as F
```

```python
N = torch.zeros((len(chars1),len(chars2)),dtype=torch.int32)
```

```python

N.shape
```

```python
for chpair, ch in zip(zip(words[:],words[1:-1]),words[2:]):
  try:
    N[c1toi[ch],c2toi[chpair]] += 1
  except:
    print(chpair, " ", ch)
```
